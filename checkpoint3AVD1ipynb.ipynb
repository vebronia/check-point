1. buatlah visualisasi data barchart


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/WLD_RTFP_country_2023-10-02 (1).csv")
df['date'] = pd.to_datetime(df['date'])
df.head()
```

# study case

1.Buatkan barchart yang menghitung total inflasi per negara

```python
print("menampilkan bar chart:")
plt.figure(figsize=(23, 10))
df.groupby('country')['Inflation'].sum().plot(kind='bar', color= 'lightblue')
plt.title('Total Inflation per Country')
plt.xlabel('Country')
plt.ylabel('Total Inflation')
plt.xticks(rotation=45)
plt.show()
```

2. membuat perkiraan tren harga pangan global pertahun!!

```python
# Histogram (Matplotlib)
plt.figure(figsize=(8, 5))
plt.hist(df['date'], bins=20, color='pink', edgecolor='black')
plt.title('jumlah harga pangan global per tahun')
plt.xlabel('Tahun')
plt.ylabel('Harga pangan')
plt.show()
```

3. Apakah terdapat korelasi yang kuat antara harga Open, High, Low, dan Close?

```python
# Heatmap (Seaborn)
plt.figure(figsize=(8, 5))
sns.heatmap(df[['Open', 'High', 'Low', 'Close']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap Korelasi Antar Variabel')
plt.show()
```
