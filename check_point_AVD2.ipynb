# Melakukan Imputasi

saya akan melakukan imputasi pada beberapa kolom, yaitu Open, Low, . Proses imputasi dilakukan karena data yang hilang pada kolom-kolom tersebut berada di bawah ambang batas 70%, sehingga memungkinkan untuk diisi menggunakan teknik seperti mean, median, atau modus.

Imputasi kolom Open

```python
df['Open'].dropna().describe()
```

```python
mean_value = df['Open'].dropna().mean()
df['Open'] = df['Open'].fillna(int(mean_value))
```

Melakukan imputasi pada kolom High

```python
df['High'].dropna().describe()
```

```python
df['High'] = df['Low'].fillna(df['Low'].dropna().mean())
```

Imputasi kolom Low

```python
df['Low'].dropna().describe()
```

```python
df['Low'] = df['Low'].fillna(df['Low'].dropna().mean())
```

imputasi pada kolom Close

```python
df['Close'].dropna().describe()
```

```python
df['Close'] = df['Close'].fillna(df['Close'].dropna().mean())
```

imputasi pada kolom Inflation



```python
df['Inflation'].dropna().describe()
```

```python
df['Inflation'] = df['Inflation'].fillna(df['Inflation'].dropna().mean())
```

#Outliers

```python
print("Pengecekan outliers:")

results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```
